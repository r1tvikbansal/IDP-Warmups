```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
sns.set()
```

```python
data = pd.read_csv('College_Data.csv')
data.head()
```

```python
sns.scatterplot(data=data, x='Grad.Rate', y='Outstate', hue='Private')
plt.savefig('grad_rate_vs_expense.png')
```

```python
sns.histplot(data=data, x='Outstate')
plt.savefig('tuition_distribution.png')
```

```python
sns.scatterplot(data=data, x='Apps', y='Accept', hue='Private')
```

```python
sns.scatterplot(data=data, x='Terminal', y='PhD', hue='Private')
```
